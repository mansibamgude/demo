```python
import numpy as np
import matplotlib.pyplot as plt

# Function to plot vectors
def Plotvec2(a, b):
    ax = plt.axes()
    ax.arrow(0, 0, *a, head_width=0.1, head_length=0.1, fc='blue', ec='blue')
    ax.arrow(0, 0, *b, head_width=0.1, head_length=0.1, fc='red', ec='red')
    plt.xlim(-2, 2)
    plt.ylim(-2, 2)
    plt.grid()
    plt.show()

# Vector subtraction u - v
u = np.array([1, 2])
v = np.array([2, 1])
subtraction_result = u - v
print("u - v:", subtraction_result)

# Multiply numpy array z with -2
z = np.array([3, 5])
z_multiplied = z * -2
print("z * -2:", z_multiplied)

# Multiply two numpy arrays element-wise
list1 = [1, 2, 3, 4, 5]
list2 = [1, 0, 1, 0, 1]
array1 = np.array(list1)
array2 = np.array(list2)
elementwise_multiplication = array1 * array2
print("Element-wise multiplication:", elementwise_multiplication)

# Convert lists to numpy arrays and plot vectors
a1 = np.array([-1, 1])
b1 = np.array([1, 1])
Plotvec2(a1, b1)
dot_product1 = np.dot(a1, b1)
print("Dot product of [-1, 1] and [1, 1]:", dot_product1)

a2 = np.array([1, 0])
b2 = np.array([0, 1])
Plotvec2(a2, b2)
dot_product2 = np.dot(a2, b2)
print("Dot product of [1, 0] and [0, 1]:", dot_product2)

a3 = np.array([1, 1])
b3 = np.array([0, 1])
Plotvec2(a3, b3)
dot_product3 = np.dot(a3, b3)
print("Dot product of [1, 1] and [0, 1]:", dot_product3)

# Explanation of dot products
print("""
The dot product for [-1, 1] and [1, 1] is zero because the vectors are orthogonal, i.e., they are perpendicular to each other.
The dot product for [1, 0] and [0, 1] is also zero because these vectors are orthogonal too.
However, the dot product for [1, 1] and [0, 1] is not zero because these vectors are not orthogonal and have a non-zero projection on each other.
""")
```
