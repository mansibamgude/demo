Exercise 1.1
Plotting a(0,50,100) and b(50,50):

```python
import matplotlib.pyplot as plt

# Data for plotting
a = [0, 50, 100]
b = [50, 50,50]

# Exercise 1.1
plt.plot(a, b, color='green')
plt.title('V graph')
plt.show()
```

Exercise 1.1 (Extension)
Plotting a(0,50,100) and b(50,50) with red color:

```python
# Exercise 1.1 (Extension)
plt.plot(a, b, color='red')
plt.title('V graph')
plt.show()
```

Exercise 1.1 (Another Extension)
Plotting a(0,50,100) and b(50,50) with different colors:

```python
# Exercise 1.1 (Another Extension)
plt.plot(a, b, color='blue', label='a part')
plt.plot([0, 50], [50, 50], color='green', label='b part')  # Only part of the graph
plt.title('L graph')
plt.legend()
plt.show()
```

Exercise 1.1 (Another Extension - Two V graphs)
Creating two "V graphs" in one graph:

```python
# Exercise 1.1 (Another Extension - Two V graphs)
plt.plot(a, b, color='green')
plt.plot([0, 50, 100], [50, 50, 50], color='green')  # Duplicate V graph
plt.title('V graph')
plt.show()
```

Exercise 1.2
Creating a 3x3 grid of plots for x(5,4,2,3,1) and y(4,2,1,3,5):

```python
# Data for Exercise 1.2
x1 = [5, 4, 2, 3, 1]
y1 = [4, 2, 1, 3, 5]

fig, axs = plt.subplots(3, 3)
fig.suptitle('3x3 Grid of Plots')

for i in range(3):
    for j in range(3):
        axs[i, j].plot(x1, y1)

plt.show()
```

Exercise 1.2 (EXTRA)
Creating a 3x3 grid of plots for two datasets:

```python
# Data for Exercise 1.2 (EXTRA)
x2 = [8, 4, 5, 3, 1]
y2 = [9, 2, 1, 1, 5]

fig, axs = plt.subplots(3, 3)
fig.suptitle('3x3 Grid of Plots')

for i in range(3):
    for j in range(3):
        axs[i, j].plot(x1, y1)
        axs[i, j].plot(x2, y2)

plt.show()
```

Exercise 1.2 (EXTRA with Colors)
Creating a 3x3 grid of plots with different colors:

```python
# Exercise 1.2 (EXTRA with Colors)
fig, axs = plt.subplots(3, 3)
fig.suptitle('3x3 Grid of Plots')

for i in range(3):
    for j in range(3):
        axs[i, j].plot(x1, y1, color='green')
        axs[i, j].plot(x2, y2, color='red')

plt.show()
```

Exercise 1.3
Creating a 2x2 grid of plots using object-oriented approach:

```python
# Data for Exercise 1.3
x3 = [2, 3, 1]
y3 = [2, 1, 3]

fig, axs = plt.subplots(2, 2)
fig.suptitle('2x2 Grid of Plots')

for ax in axs.flat:
    ax.plot(x3, y3)

plt.show()
```

Exercise 1.6
Plotting two or more lines with different styles:

```python
# Exercise 1.6
plt.plot(x1, y1, linestyle='--', label='Dashed Line')
plt.plot(x2, y2, linestyle='-', label='Solid Line')
plt.legend()
plt.show()
```
